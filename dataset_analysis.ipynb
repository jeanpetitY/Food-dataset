```python
from transformers import AutoImageProcessor, AutoModelForImageClassification
from PIL import Image
import torch

# ---- Charger ton mod√®le pr√©f√©r√© ----
model_name = "prithivMLmods/Food-101-93M"

# AutoModelForImageClassification est plus g√©n√©rique que SiglipForImageClassification
model = AutoModelForImageClassification.from_pretrained(model_name)
processor = AutoImageProcessor.from_pretrained(model_name)

# ---- Activer GPU si dispo ----
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# ---- R√©cup√©rer le mapping id -> label ----
id2label = model.config.id2label

def predict_food_name(image_path: str) -> str:
    """
    Pr√©dit le nom du plat √† partir du chemin d'une image.
    """
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        logits = model(**inputs).logits
        pred_id = logits.argmax(-1).item()

    predicted_label = id2label[pred_id]
    return predicted_label

# ---- Exemple d‚Äôutilisation ----
if __name__ == "__main__":
    image_path = "baby_back_ribs.jpg"  # ton image
    result = predict_food_name(image_path)
    print(f"üçΩÔ∏è Predicted food: {result}")
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

"""
Fine-tune BEiT or ViT on a balanced image classification dataset.
Each subfolder under /train and /val represents a class.
Author: Jean Petit BIKIM (with GPT-5 support)
"""

import argparse
import json
import os
from dataclasses import dataclass
from typing import Any, Dict

import numpy as np
import random
from PIL import Image
from datasets import load_dataset, DatasetDict
import evaluate
from sklearn.metrics import f1_score
from transformers import (
    AutoImageProcessor,
    AutoModelForImageClassification,
    TrainingArguments,
    Trainer,
    DefaultDataCollator,
)


# -------------------------- Utility Functions --------------------------
def seed_everything(seed: int = 42):
    """Ensure deterministic behavior for reproducibility."""
    random.seed(seed)
    np.random.seed(seed)


def infer_splits(data_dir: str):
    """Return paths to train and validation directories if they exist."""
    train_dir = os.path.join(data_dir, "train")
    val_dir = os.path.join(data_dir, "test")
    return train_dir, val_dir if os.path.isdir(val_dir) else (train_dir, None)


# -------------------------- Image Preprocessing --------------------------
@dataclass
class ImageTransform:
    """Applies only normalization and resizing using the model processor."""

    processor: AutoImageProcessor
    train: bool = True  # kept for compatibility; not used here

    def __call__(self, examples: Dict[str, Any]) -> Dict[str, Any]:
        """Convert images to tensors using the processor (no augmentation)."""
        images = [
            x.convert("RGB") if isinstance(x, Image.Image) else Image.open(x).convert("RGB")
            for x in examples["image"]
        ]
        inputs = self.processor(images=images, return_tensors="pt")
        result = dict(inputs.items())
        result["labels"] = examples["label"]
        return result


# -------------------------- Metrics --------------------------
accuracy = evaluate.load("accuracy")


def compute_metrics(eval_pred):
    """Compute accuracy and macro F1 score for evaluation."""
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    acc = accuracy.compute(predictions=preds, references=labels)["accuracy"]
    f1_macro = f1_score(labels, preds, average="macro")
    return {"accuracy": acc, "f1_macro": f1_macro}


# -------------------------- Main Training Function --------------------------
def main():
    parser = argparse.ArgumentParser(description="Fine-tune BEiT or ViT on a folder dataset")
    parser.add_argument("--data_dir", type=str, required=True, help="Root directory containing train/ and optionally val/")
    parser.add_argument("--model_name", type=str, default="google/vit-base-patch16-224",
                        help="Model name (e.g., google/vit-base-patch16-224 or microsoft/beit-base-patch16-224-pt22k-ft22k)")
    parser.add_argument("--output_dir", type=str, default="./outputs")
    parser.add_argument("--image_size", type=int, default=400)
    parser.add_argument("--epochs", type=int, default=10)
    parser.add_argument("--batch_size", type=int, default=16)
    parser.add_argument("--lr", type=float, default=5e-5)
    parser.add_argument("--weight_decay", type=float, default=0.05)
    parser.add_argument("--warmup_ratio", type=float, default=0.05)
    parser.add_argument("--val_ratio", type=float, default=0.1, help="Used if no val folder is provided")
    parser.add_argument("--seed", type=int, default=42)
    parser.add_argument("--fp16", action="store_true", help="Enable mixed precision (AMP) if supported")
    parser.add_argument("--gradient_accumulation", type=int, default=1)
    parser.add_argument("--freeze_backbone", action="store_true", help="Freeze all layers except the classification head")
    args = parser.parse_args()

    # Set random seed
    seed_everything(args.seed)

    # Identify dataset splits
    train_dir, val_dir = infer_splits(args.data_dir)

    # Load dataset from image folders
    if val_dir is not None:
        ds = DatasetDict({
            "train": load_dataset("imagefolder", data_dir=train_dir, split="train"),
            "validation": load_dataset("imagefolder", data_dir=val_dir, split="train"),
        })
    else:
        full = load_dataset("imagefolder", data_dir=train_dir, split="train")
        ds_split = full.train_test_split(test_size=args.val_ratio, seed=args.seed)
        ds = DatasetDict({"train": ds_split["train"], "validation": ds_split["test"]})

    # Extract label mapping
    labels = ds["train"].features["label"].names
    id2label = dict(enumerate(labels))
    label2id = {l: i for i, l in enumerate(labels)}

    # Load processor (resize, normalize, etc.)
    processor = AutoImageProcessor.from_pretrained(
        args.model_name,
        size={"shortest_edge": args.image_size} if "beit" in args.model_name else args.image_size
    )

    # Apply transforms (no augmentation)
    ds["train"].set_transform(ImageTransform(processor, train=True))
    ds["validation"].set_transform(ImageTransform(processor, train=False))

    # Load pre-trained model
    model = AutoModelForImageClassification.from_pretrained(
        args.model_name,
        num_labels=len(labels),
        id2label=id2label,
        label2id=label2id,
        ignore_mismatched_sizes=True,
    )

    # Optionally freeze the backbone
    if args.freeze_backbone:
        for name, param in model.named_parameters():
            param.requires_grad = "classifier" in name or "head" in name
    # Define training arguments
    training_args = TrainingArguments(
        output_dir=args.output_dir,
        num_train_epochs=args.epochs,
        per_device_train_batch_size=args.batch_size,
        per_device_eval_batch_size=args.batch_size,
        learning_rate=args.lr,
        weight_decay=args.weight_decay,
        warmup_ratio=args.warmup_ratio,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        metric_for_best_model="f1_macro",
        greater_is_better=True,
        logging_steps=50,
        fp16=args.fp16,
        gradient_accumulation_steps=args.gradient_accumulation,
        report_to="none",
        dataloader_num_workers=4,
    )

    # Data collator (for batching)
    collator = DefaultDataCollator()

    # Initialize Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=ds["train"],
        eval_dataset=ds["validation"],
        tokenizer=processor,
        data_collator=collator,
        compute_metrics=compute_metrics,
    )

    # Train the model
    trainer.train()

    # Save model and processor
    os.makedirs(args.output_dir, exist_ok=True)
    model.save_pretrained(args.output_dir)
    processor.save_pretrained(args.output_dir)
    with open(os.path.join(args.output_dir, "label_mapping.json"), "w", encoding="utf-8") as f:
        json.dump({"id2label": id2label, "label2id": label2id}, f, indent=4, ensure_ascii=False)

    # Evaluate final performance
    metrics = trainer.evaluate()
    with open(os.path.join(args.output_dir, "eval_metrics.json"), "w") as f:
        json.dump(metrics, f, indent=2)

    print("‚úÖ Training completed successfully.")
    print("üìä Final evaluation metrics:", metrics)


if __name__ == "__main__":
    main()
```

```python
import json
import matplotlib.pyplot as plt

def load_training_log(jsonl_path):
    """Charge un fichier .jsonl et retourne les epochs, losses et learning rates."""
    epochs, losses, lrs = [], [], []
    with open(jsonl_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            try:
                record = json.loads(line.replace("'", '"'))  # Corrige les quotes simples
                if "epoch" in record and "loss" in record:
                    epochs.append(float(record["epoch"]))
                    losses.append(float(record["loss"]))
                    lrs.append(float(record.get("learning_rate", 0.0)))
            except json.JSONDecodeError:
                continue
    return epochs, losses, lrs


# === üîß Chemins vers tes fichiers JSONL ===
beit_file = "finetuning/metrics/cv/beit.jsonl"
vit_file = "finetuning/metrics/cv/vit.jsonl"

# === Chargement des donn√©es ===
epochs_beit, losses_beit, lrs_beit = load_training_log(beit_file)
epochs_vit, losses_vit, lrs_vit = load_training_log(vit_file)

# === R√©cup√©ration du dernier learning rate pour chaque mod√®le ===
lr_beit_last = lrs_beit[-1] if lrs_beit else 0
lr_vit_last = lrs_vit[-1] if lrs_vit else 0

# ===Trac√© des courbes ===
plt.figure(figsize=(10, 6))
plt.plot(epochs_beit, losses_beit, label=f"BEiT (lr={lr_beit_last:.2e})", color="blue", linewidth=2)
plt.plot(epochs_vit, losses_vit, label=f"ViT (lr={lr_vit_last:.2e})", color="orange", linewidth=2)

# plt.title("Comparaison  entre BEiT et ViT", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
